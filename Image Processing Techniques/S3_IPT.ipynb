```python
img= cv2.imread('Osteosarcoma_01_25Sigma_noise.tif',0)
```

```python
plt.imshow(img,cmap='gray')
```

```python
guassian_blur= cv2.GaussianBlur(img,(5,5),0)
```

```python
plt.imshow(guassian_blur,cmap='gray')
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(img,cmap='gray')
plt.title('Original')
plt.subplot(1,2,2)
plt.imshow(guassian_blur,cmap='gray')
plt.title('Blurred')
plt.show
```

```python
img_sharp=np.array([[0,-1,0],
                    [-1,5,1],
                    [0,-1,0]])
out_im=cv2.filter2D(img,-1,img_sharp)
plt.imshow(out_im,cmap='gray')
```
