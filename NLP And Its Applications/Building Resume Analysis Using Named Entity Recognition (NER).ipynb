```python
import spacy

# Load the pre-trained spaCy model for English
nlp = spacy.load("en_core_web_sm")

# Sample resume text data
resume_text = """
John Doe
123 Main Street
Anytown, USA 12345
(123) 456-7890
john.doe@email.com

Objective:
To secure a challenging position as a Data Scientist in a reputable organization to expand my learnings, knowledge, and skills.

Experience:
Data Scientist
ABC Company, Anytown, USA
June 2018 - Present
- Develop and maintain machine learning models to analyze customer data.
- Collaborate with cross-functional teams to design data-driven solutions.

Junior Data Analyst
XYZ Corporation, Anytown, USA
May 2016 - May 2018
- Analyzed sales data to identify trends and patterns.
- Generated reports and visualizations to support business decisions.

Education:
Master of Science in Data Science
University of Anytown, USA
September 2014 - May 2016

Bachelor of Science in Computer Science
University of Anytown, USA
September 2010 - May 2014

Skills:
- Python, R, SQL
- Machine Learning, Data Analysis, Data Visualization
- TensorFlow, Keras, Scikit-Learn
"""

# Process the resume text with spaCy
doc = nlp(resume_text)

# Extract named entities
print("Named Entities in the Resume:")
for ent in doc.ents:
    print(f"{ent.text}: {ent.label_}")

# Example: Filtering for specific types of entities (e.g., PERSON, ORG, GPE, DATE)
print("\nFiltered Entities:")
for ent in doc.ents:
    if ent.label_ in ["PERSON", "ORG", "GPE", "DATE"]:
        print(f"{ent.text}: {ent.label_}")

# Example: Extracting specific sections like education or experience
education_section = resume_text.split("Education:")[1].split("Skills:")[0].strip()
print("\nExtracted Education Section:")
print(education_section)

experience_section = resume_text.split("Experience:")[1].split("Education:")[0].strip()
print("\nExtracted Experience Section:")
print(experience_section)
```
