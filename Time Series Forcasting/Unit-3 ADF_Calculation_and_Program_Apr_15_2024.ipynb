# Augmented Dickey-Fuller (ADF) Test



The Augmented Dickey-Fuller (ADF) test is a statistical test used to determine whether a unit root is present in a time series dataset. A unit root indicates that a time series variable is non-stationary, meaning its statistical properties, such as mean and variance, are not constant over time. The presence of a unit root implies that the series has a stochastic trend and might exhibit random behavior, making it difficult to make reliable predictions or draw meaningful conclusions from the data.

Here's a breakdown of the significance, coverage, and validity of the ADF test:

###Significance:

The ADF test helps determine whether a given time series is stationary or non-stationary.
If the null hypothesis of the ADF test is rejected, it suggests that the time series is stationary, meaning it lacks a unit root. In other words, the series exhibits stable statistical properties over time.
On the other hand, if the null hypothesis is not rejected, it suggests that the time series is non-stationary, indicating the presence of a unit root.

###Coverage:

The ADF test is widely used in econometrics, finance, and other fields where time series data analysis is common.
It can be applied to various types of time series data, including economic indicators, financial market data, weather patterns, and more.
The test can be employed for both univariate and multivariate time series analysis.

###Validity:

The ADF test is based on the Dickey-Fuller test, which was developed to test for the presence of a unit root in a time series.
The ADF test extends the Dickey-Fuller test by incorporating additional terms to account for potential autocorrelation and serial correlation in the data.
The validity of the ADF test relies on certain assumptions, such as the presence of weak dependence and the absence of structural breaks in the data.
It's important to note that while the ADF test can provide valuable insights into the stationarity of a time series, it is not immune to certain limitations, such as sensitivity to sample size and model specification.




The ADF test is a significant tool for analyzing time series data, with broad coverage across various disciplines. However, its validity depends on certain assumptions and considerations, and it should be used judiciously in conjunction with other analytical techniques for robust data analysis and interpretation.

```python
import numpy as np
import pandas as pd
from statsmodels.tsa.stattools import adfuller

# Generate synthetic non-stationary time series data
np.random.seed(42)
n_samples = 100
non_stationary_data = np.cumsum(np.random.randn(n_samples))

# Define a function to perform the ADF test and print results
def adf_test(data):
    result = adfuller(data)
    print('ADF Statistic:', result[0])
    print('p-value:', result[1])
    print('Critical Values:')
    for key, value in result[4].items():
        print(f'   {key}: {value}')

# Perform ADF test on the non-stationary data
print("ADF Test Results for Non-Stationary Data:")
adf_test(non_stationary_data)

# Generate synthetic stationary time series data
stationary_data = np.random.randn(n_samples)

# Perform ADF test on the stationary data
print("\nADF Test Results for Stationary Data:")
adf_test(stationary_data)
```

```python
def adf_test(data):
    # Calculate the number of lags for the test
    n = len(data)
    k = int(np.ceil(12 * np.power(n / 100, 1/4)))

    # Compute first differences
    data_diff = np.diff(data)

    # Compute lagged differences
    lagged_diff = np.zeros((n - k, k))
    for i in range(k):
        lagged_diff[:, i] = data_diff[i:n-k+i]

    # Prepare the regression matrix
    X = np.column_stack((np.ones(n - k), data_diff[k - 1:-1], lagged_diff))
    y = data_diff[k:]

    # Calculate the coefficients
    beta = np.linalg.inv(X.T @ X) @ X.T @ y

    # Compute the residuals
    residuals = y - X @ beta

    # Compute the residual sum of squares
    rss = np.sum(residuals**2)

    # Compute the total sum of squares
    tss = np.sum((y - np.mean(y))**2)

    # Calculate the test statistic
    test_statistic = (rss / k) / (tss / (n - k - 1))

    # Calculate the p-value (approximation for ADF test)
    p_value = 1 - 0.348 * np.log(n)**2 / n + 1.912 * np.log(n) / n

    # Define the critical values for different significance levels
    critical_values = {
        1: -3.43,
        5: -2.86,
        10: -2.57
    }

    # Determine the result based on the test statistic and critical values
    result = ''
    if test_statistic < critical_values[1]:
        result = "Stationary (significant at 1% level)"
    elif test_statistic < critical_values[5]:
        result = "Stationary (significant at 5% level)"
    elif test_statistic < critical_values[10]:
        result = "Stationary (significant at 10% level)"
    else:
        result = "Non-Stationary"

    # Return the test results
    return {
        'ADF Statistic': test_statistic,
        'p-value (approximation)': p_value,
        'Result': result
    }

```

***Simple Operators***

We shall employ extensively the backward shift operator $B$, which is defined by $Bx_t = x_{t-1}$; it leads to $B^mz_t = x_{t-m}$. The inverse operation is
performed by the forward shift operator $F = B^{-1}$ given by $Fx_t = x_{t+1}$; hence
$F^m x_t = x_{t+m}$. Another important operator is the backward difference operator, $\nabla$,
defined by $\nabla x_t =z_t - x_{t-1}$ . This can be written in terms of $B$, since since
$\nabla x_t = x_t - x_{t-1} = (1 - B)x_t$

A seasonal difference followed by a first difference can be written as
\begin{align*}
	(1-B)(1-B^m)y_t &= (1 - B - B^m + B^{m+1})y_t \\
	&= y_t-y_{t-1}-y_{t-m}+y_{t-m-1},
\end{align*}

**Autoregressive Models**

A stochastic model that can be extremely useful in the representation of certain practically occurring series is the auto-regressive model. In this model, the current value of the process is expressed as a finite, linear aggregate of previous values of the process and a random shock at. Let us denote the values of a process at equally spaced times $t,t - 1, t - 2, \dots$ by $y_{t},y_{t-1},y_{t-2}, \dots$. Also let $\bar{y}_{t} = {y}_{t} - \mu$ be the series of deviations from $\mu$.


\begin{equation}
	\label{eqn-arp}
	y_{t} = c + \phi_{1}y_{t-1} + \phi_{2}y_{t-2} + \dots + \phi_{p}y_{t-p} + \varepsilonlon_{t}
\end{equation},
\begin{equation}
	\label{eqn-arp_mu}
	\overline{y}_{t} = c + \phi_{1}\overline{y}_{t-1} + \phi_{2}\overline{y}_{t-2} + \dots + \phi_{p}\overline{y}_{t-p} + \varepsilon_{t}
\end{equation}

If we define an auto-regressive operator of order $p$ in terms of the backward shift operator $B$ by
\begin{equation}
	\phi(B) = 1 - \phi_1B - \phi_2B^2 - \cdots - \phi_pB^p
\end{equation}
the auto-regressive model \eqref{eqn-arp} may be written economically as $\phi(B)y_t = \varepsilon_{t}$
The model contains $p + 2$ unknown parameters $\mu, \phi_1, \phi_2, \dots , \phi_p, \sigma_\varepsilon^2$, which in practice have to be estimated from the data. The additional parameter $\sigma_\varepsilon^2$ is the variance of the white noise process $\varepsilon_{t}$.






**Unit Root: **

Unit root is one of the special characteristics of a time series which tells about the non-stationarity of the time series. Technologically tells, a unit root is said to exist in a time series of the value of $\alpha = 1$ in the below equation.
$$x_t=\alpha  x_{t-1} +\beta y_e +\varepsilon  $$
where, $x_t$ is the value of the time series at the time $t$ and $y_e$ is an exogenous variable (a separate explanatory variable, which is also a time series).

**ADF Testing :**

The Augmented Dickey Fuller test (ADF) is a modification of the Dickey-Fuller (DF) unit root. Dickey-Fuller used a combination of $t$-statistics and $F$-statistics to detect the presence of a unit root in time series.

ADF test in pairs trading is done to check the co-integration between two stocks (presence of unit root).

If there is a unit root present in the time series, it implies that the time series is non-stationary and the stocks are not co-integrated. Hence, stocks cannot be traded together.
Alternatively, if the null hypothesis gets rejected and the stocks show co-integration; it implies that the time series is stationary and the stocks can be traded.


The main point of conducting an ADF test in pairs trading strategy is to ensure if the pair of stocks considered in the strategy are stationary or not.
Hence, for the pair of stocks to be traded in the pairs trading strategy, it is required that the time series is stationary. A stationary time series makes effective and precise predictions.
Also, a stationary time series means that the pair of stocks is co-integrated and can be traded together by generating trading signals.

Similar to regular hypothesis testing methods, Augmented Dickey-Fuller test is also based on two hypothesis.  

The null hypothesis states that there exists a unit root in the time series and is non-stationary.

The alternative hypothesis states that there exists no unit root in the time series and is stationary or trend stationary.

Let us derive the actual structure of the Dickey-Fuller Test.


$$x_t= c+ \beta_t+\alpha x_{t-1}+\phi\Delta x_{t-1}+\varepsilon_t$$
where,
$x_t=$ value in the time series at time $t$ \textbf{or} time series with lag-1 and
$\Delta x_{t} =$ first difference of the series at time $(t-1)$

Now, we will see the formula for Augmented Dickey Fuller test, and it goes as follows

$$x_t= c+ \beta_t+\alpha x_{t-1}+\phi_1\Delta x_{t-1}+\phi_2\Delta x_{t-2}+\dots+\phi_p\Delta x_{t-p}$$
One can see that the formula for ADF is the similar as of DF with the only difference being the addition of differencing terms representing a larger time series.

The test is performed by rewriting the model as follows;
\begin{eqnarray*}
x_t- x_{t-1}&=&(\phi_1-1)x_{t-1}+\phi_2x_{t-2}+\phi_3 x_{t-3}+\varepsilon_t \\
x_t- x_{t-1}&=&(\phi_1-1)x_{t-1}+\left(\phi_2+\phi_3\right)x_{t-2}+\phi_3 \left(x_{t-3}-x_{t-2}\right)+\varepsilon_t \\
x_t- x_{t-1}&=&(\phi_1+\phi_2+\phi_3-1)x_{t-1}+\left(\phi_2+\phi_3\right)\left(x_{t-2}-x_{t-1}\right)+\phi_3 \left(x_{t-3}-x_{t-2}\right)+\varepsilon_t \\
x_t&=&(\pi)x_{t-1}+\left(c_1\right)\left(x_{t-1}\right)+c_2 \left(x_{t-2}\right)+\varepsilon_t \end{eqnarray*}
where,
\begin{eqnarray*}
	\pi&=&\phi_1+\phi_2+\phi_3-1=\phi-1\\
	c_1&=&-\left(\phi_2+\phi_3\right)\\
	c_2&=&-\phi_3
\end{eqnarray*}


The unit root null hypothesis against the stationary alternative corresponds to
$H_0\, :\, \phi = 1$ against $H_A$ or $H_1\, :\, \phi < 1$\\
Alternatively, for the transformed model equations with $\pi$, we have the hypothesis of the form $H_0\, :\, \pi = 0$ against $H_A$ or $H_1\, :\, \pi < 0$

The $t-$test for $H_0$ is denoted as the Augmented Dickey-Fuller (ADF) test which is given by the test


\begin{eqnarray*}
ADF &=& \frac{\widehat{\phi} -1}{SE\left(\widehat{\phi}\right)}=\frac{\widehat{\pi} -0}{SE\left(\widehat{\pi}\right)}
\end{eqnarray*}

###Least Square Estimation of AR Model
Consider an autoregressive model of order $p$ driven by white noise.

\begin{equation*}
	y_n = \sum_{k=1}^{p} a_ky_{n-k} + \varepsilon_n
\end{equation*}where, $\varepsilon$
is a Gaussian white noise with zero mean and some variance $\sigma_{\varepsilon}^2$.

Let us assume that we have $N+1$ sample items of this AR process and we are interested in estimating the parameter $a_k$. We can arrange the data into a set of $M=N+1-p$ linear equations,


\begin{eqnarray*}
	\begin{bmatrix}
	y_{0} & y_{1} & \cdots & y_{p-1} \\
	y_{1} & y_{2} & \cdots & y_{p} \\
	y_{2} & y_{3} & \cdots & y_{p+1} \\
	% y_{p+1} & y_{p} & \cdots & y_{2} \\
	\vdots & \vdots & \ddots & \vdots \\
	y_{N-p} & y_{N-p+1} & \cdots & y_{N-1}
\end{bmatrix} \begin{bmatrix}
	a_p \\
	a_{p-1} \\
	\vdots \\
	a_1
\end{bmatrix} &=& \begin{bmatrix}
	y_{p}\\
	y_{p+1}\\
	y_{p+2}\\
	\vdots\\
	y_{N}
\end{bmatrix} \\\\%%\longrightarrow vA = v
\left(\begin{bmatrix}
%%v_n^T
	v_{N-p, M} & v_{N-p+1, M} &  
	\ldots & v_{N-1, M}
\end{bmatrix}\right)A &=& v_{N, M}\\\\
\left(Y_{N,M}\right)A&=&v_{N, M}
\end{eqnarray*}

where $v_{k, M}$ is a column vector whose elements are the past $M$ of $y_{n}$ starting from the instant $k$. $y_{k-M+1},y_{k}$ are the first and last elements of the vector, respectively.

From the above we see that 	$Y_{N,M}$ consists of the column vectors 	$v_{N-p, M}, v_{N-p+1, M}, \dots v_{N-1, M}$.



The least-squares solution to this problem is the following,

\begin{equation*}
	\widehat{A} = \left(Y_{N,M}^TY_{N,M}\right)^{-1}Y_{N,M}^Tv_{N,M}
\end{equation*}

Post-multiplying $v_{N,M}$ by the pseudo-inverse will provide the least square estimate of $A$.

##White Noising (White Noise Model) using $\widehat{a}$
Once $\widehat{a}$ is obtained, then the signal $y_{n}$ can be whitened by passing it through the following moving average filter.

\begin{equation*}
	w_n = y_n - \sum_{k=1}^{p} \hat{a}_ky_{n-k}
\end{equation*}


$w_{n}$ would be the out of this moving average filter, and $w_{n}$
will be a white noise. The input to this moving average filter the measured time series $y_{n}$.

3. With the help of estimated Parameter, we have to perform ADF test.
Hypothesis for the the test is

**Null Hypothesis $H_0$ :** Time is not stationary (There seems to be Unit Root)

**Alternative Hypothesis $H_1$ :** Time is stationary (There seems to be no Unit Root)

\begin{equation*}
  ADF=\frac{\widehat{\phi}-1}{\sqrt{\Big(\frac{\sum_{t=2}^{n}(y_t-\widehat{\phi}y_{t-1})^2}{n-3}\Big)/(\sum_{t=2}^n(y_{t-1}-\bar{y})^2)}}
\end{equation*}

Since we have taken $\widehat{a}$ as our estimate, we rewrite the formula for our convenient as follows:
\begin{equation*}
  ADF=\frac{\widehat{a}-1}{\sqrt{\Big(\frac{\sum_{t=2}^{n}(y_t-\widehat{a}y_{t-1})^2}{n-3}\Big)/(\sum_{t=2}^n(y_{t-1}-\bar{y})^2)}}
\end{equation*}

$$ \text{adj}(X^TX) = \begin{bmatrix}

```python
x=df1["Year"]
x1=df1["Sales"];   ####x2=df1.X2;x3=df1.X3;


# Generating AR Model
def AR_p(p, a, eparam, eps):
    N = len(eps)
    me, ve = eparam
    # Generate the filtered signal
    y = np.zeros(N)
    for i in range(p, N):
        y[i] = eps[i]
        for k in range(p):
            y[i] += a[k] * y[i - k - 1]
    return y
```

```python
def AR_param(p, y):
    N = len(y)
    ymat = np.zeros((N - p, p))
    yb = np.zeros((N - p - 1, 1))
    for c in range(p, 0, -1):
        ymat[:, p-c] = y[p-c:-c]
    yb = y[p:]
    return np.matmul(np.linalg.pinv(ymat), yb)[::-1]
```

```python
# Implementation of stationary white noise filter.
def w_n(y, acap):
    N = len(y)
    p = len(acap)
    w = np.zeros(N)
    for i in range(N):
        w[i] = y[i]
        for k in range(p):
            if i - k - 1 >= 0:
                w[i] += - acap[k] * y[i - k - 1]
    return w
```

```python
from math import log
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# calculate aic for regression
def calculate_aic(n, mse, num_params):
	aic = n * log(mse) + 2 * num_params
	return aic

# generate dataset
X, y = make_regression(n_samples=100, n_features=2, noise=0.1)
# define and fit the model on all data
model = LinearRegression()
model.fit(X, y)
# number of parameters
num_params = len(model.coef_) + 1
print('Number of parameters: %d' % (num_params))
# predict the training set
yhat = model.predict(X)
# calculate the error
mse = mean_squared_error(y, yhat)
print('MSE: %.3f' % mse)
# calculate the aic
aic = calculate_aic(len(y), mse, num_params)
print('AIC: %.3f' % aic)
```
