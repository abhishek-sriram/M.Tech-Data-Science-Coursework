#ARIMA

A time series $\{X_t\}$ is said to follow an integrated autoregressive moving average model if the $d^{\text{th}}$ difference $W_t = ∇^dX_t=(1-B)^dX_t$ is a stationary ARMA process. If $\{W_t\}$ follows an $ARMA(p,q)$ model, we say that $\{X_t\}$ is an $ARIMA(p,d,q)$ process. In general, to avoid much complications on calculation and for checking the correctness of the calculations, we can usually take $d = 1$ or at most 2 for practical purposes.

Consider then an $ARIMA(p,1,q)$ process with $W_t = X_t − X_{t − 1}$, we have

$W_t = \phi_1 W_{t-1} + \dots + {\phi_p}W_{t-p}+\varepsilon_t - \{\theta_1 \varepsilon_{t-1} + \dots + \theta_{q} \varepsilon_{t-q}\}$

where $\varepsilon_t \sim WN(0, \sigma^2).$

In terms of the observed data series,

$X_t − X_{t − 1} = \phi_1 \left[X_{t-1} − X_{t − 2}\right] + \dots + {\phi_p}\left[X_{t-p} − X_{t − p-1}\right]+\varepsilon_t - \{\theta_1 \varepsilon_{t-1} + \dots + \theta_{q} \varepsilon_{t-q}\}$

It can be written in terms of backshift operator as,
$\phi(B) (1-B)^d (X_t) = \theta (B) \varepsilon_t $ with $d=1$

#SARIMA

Suppose that $\{X_t\}$ exhibits a seasonal trend, in the sense that $X_t \sim X_{t-s} \sim
Y_{t-2s} \dots $.
Then it not only depends on $X_{t-1}, X_{t-2}, \dots,$ but also $X_{t-s},X_{t-2s},\dots$.
To model this, consider

$\phi(B) \Phi_P(B^s) (1-B)^d(1-B^s)^D (X_t) = \theta (B) \Theta_Q(B^s)\varepsilon_t $

is a $SARIMA(p,d,q)\times(P,D,Q)_s$ model with $s$ as a seasonal component(period).
where
\begin{eqnarray}
	\phi(B) &=& 1 - \phi_1B - \phi_2B^2 - \cdots - \phi_pB^p \\
	\theta(B) &=& 1 - \theta_1B - \theta_2B^2 - \cdots - \theta_qB^q \\
  \Phi(B^s) &=& 1 - \Phi_1B^s - \Phi_2\left(B^s\right)^2 - \cdots - \Phi_p\left(B^s\right)^P \\
  \Theta(B^s) &=& 1 - \Theta_1B^s - \Theta_2\left(B^s\right)^2 - \cdots - \Theta_q\left(B^s\right)^Q
\end{eqnarray}


Such $\{X_t\}$ is usually denoted by $SARIMA (p, d, q) \times (P,D,Q)_s$ or $SARIMA (p, d, q) \times (P,D,Q,s)$.

Of course, we could expand the right-hand side of above equation and express  $\{X_t\}$ in terms of a higher-order ARMA model (see the following example).
But we prefer the $SARIMA$ format, due to its natural interpretation.


Let us consider the structure of an $SARIMA(1,0,0) \times(0,1,1)_{12}$ time series $\{X_t\}$, which is expressed as
	\begin{equation*}
		(1-\phi B)(1-B^{12})X_t=\varepsilon_t-\theta \varepsilon_{t-12}, \\
    (1-B^{12}-\phi B + \phi B^{13})X_t=\varepsilon_t-\theta \varepsilon_{t-12}
	\end{equation*}
so that
\begin{equation*}
	X_t=X_{t-12}+\phi (X_{t-1}-X_{t-13})+X_t-\theta X_{t-12}.
\end{equation*}
Notice that $X_t$ depends on $Y_{t-12},Y_{t-1},Y_{t-13}$ as well as $\varepsilon_{t-12}$.
If $\{X_t\}$  represents monthly observations over a number of years, we can tabulate the data using two-way ANOVA as follows:

## Stationarity/Causal conditions of an AR Model/Process

For an AR(1) model, $(1 − \phi B)X_t = \varepsilon_t$, to be causal, the root of $\phi(z) = 1 − \phi z$ must lie outside of the unit circle. In this case, $\phi(z) = 0$ when $z = 1/\phi$, so it is easy to go from the causal requirement on the root, $\left|\frac{1}{\phi}\right| > 1$, to a requirement on the parameter, $\left|\phi\right| < 1$. It is not so easy to establish this relationship for higher order
models.


Consider a mean-centred AR(2) process
\begin{equation}
X_t=\phi_1X_{t-1}+\phi_2X_{t-2}+\varepsilon_t
\end{equation}
where $\varepsilon_t$ is the standard white noise process.

The above AR(2) model can be rewritten in terms of backshift operator to the form $(1 − \phi_1B − \phi_2B^2)X_t = \varepsilon_t$, is causal when the two roots of $\phi(z) = 1 − \phi1z − \phi_2z^2$ lie outside of the unit circle.
<!-- Using the quadratic formula, this requirement can be written as -->

Consider the equation

$\lambda^2-\phi_1\lambda-\phi_2=0$

If $z$ is a root of the "standard" characteristic equation $1−ϕ_1z−ϕ_2z^2=0$

 and setting $z^{−1}=λ$, the display obtains from rewriting the standard one as follows:

\begin{eqnarray*}
1-\phi_1 z-\phi_2 z^2&=&0\\
\Rightarrow z^{-2}-\phi_1 z^{-1}-\phi_2 &=&0\\
\Rightarrow \lambda^2-\phi_1\lambda -\phi_2 &=&0
\end{eqnarray*}



Hence, an alternative condition for stability of an AR(2)
 is that all roots of the first display are inside the unit circle,

 $|z|>1 \Leftrightarrow |\lambda|=|z^{-1}|<1$



We use this representation to derive the stationarity triangle of an AR(2) process, that is that an AR(2) is stable if the following three conditions are met:

1. $\phi_2 <1 +\phi_1$
1. $\phi_2< 1-\phi_1$
1. $\phi_2>-1$

Recall that you can write the roots of the first display (if real) as

$\lambda_{1,2}=\frac{\phi_1\pm\sqrt{\phi_1^2+4\phi_2}}{2}$
to find the first two conditions.

Then, the AR(2)  is stationary iff $|λ|<1$, hence (if the $λ_i, i=1,2$
 are real):
\begin{eqnarray*}
-1<\frac{\phi_1\pm\sqrt{\phi_1^2+4\phi_2}}{2}&<&1\\
\Rightarrow    -2<\phi_1\pm\sqrt{\phi_1^2+4\phi_2}&<&2
\end{eqnarray*}


The larger of the two $λ_i, i=1,2$ is bounded by $\phi_1+\sqrt{\phi_1^2+4\phi_2}<2$, or:

\begin{eqnarray*}
\phi_1+\sqrt{\phi_1^2+4\phi_2}&<&2\\
\Rightarrow \sqrt{\phi_1^2+4\phi_2}&<&2 - \phi_1\\
\Rightarrow \phi_1^2+4\phi_2&<&(2 - \phi_1)^2\\
\Rightarrow \phi_1^2+4\phi_2&<&4 - 4\phi_1+\phi_1^2\\
\Rightarrow \phi_2&<&1 - \phi_1
\end{eqnarray*}


Analogously, we find that $ϕ_2<1+ϕ_1$.

If $λ_i, i=1,2$ is complex, then $ϕ_1^2<−4ϕ_2$ and so

$\lambda_{1,2} = \phi_1/2\pm i\sqrt{-(\phi_1^2+4\phi_2)}/2.$

The squared modulus of a complex number is the square of the real plus the square of the imaginary part. Hence,

$\lambda^2 = (\phi_1/2)^2 + \left(\sqrt{-(\phi_1^2+4\phi_2)}/2\right)^2 = \phi_1^2/4-(\phi_1^2+4\phi_2)/4 = -\phi_2.$


This is stable if $|λ|<1$, hence if $−ϕ_2<1$  or $ϕ_2>−1$, as was to be shown. (The restriction $ϕ_2<1$ resulting from $ϕ_2^2<1$ is redundant in view of $ϕ_2<1+ϕ_1$ and $ϕ_2<1−ϕ_1.$)

- Pure MA models are always stationary (since they contain no AR terms).
- Pure MA models may or may not be invertible.
- Pure AR models are always invertible (since they contain no MA terms).
- Pure AR models may or may not be stationary.

This translates to checking the following:

- If **all** the roots of $\phi(z) = 0$ are outside the unit circle, the process is stationary;
- If **all** the roots of $\theta(z) = 0$ are outside the unit circle, the process is invertible;


Recall that

**Causality** is about writing $X_t$ in terms of the driving noise
process $\{\varepsilon_t, \varepsilon_{t−1}, \varepsilon_{t−2}, \dots \}$.

**Invertibility** is about writing $\varepsilon_t$ in terms of $\{X_t, X_{n−1}, X_{n−2}, \dots \}$.

**Stationary:** $\phi$(.) roots are not on the unit circle.

**Casuality:** $\phi$(.) roots are outside unit circle.

**Invertibility:** $\theta$(.) roots are outside the unit circle.

- (A) $1 + 0.5x = 0$
- (B) $1 + 1.3x - 0.4 x^2 = 0$
- (E) $1 + 1.3x = 0$

We will not repeat the manual calculations and instead immediately look at the results via `Python`:

```python
print("(A)" + str(np.poly1d([0.5, 1], variable = 'x')))
print("Roots: " + str(np.poly1d([0.5, 1], variable = 'x').roots))
```

```python
print("(B)\n" + str(np.poly1d([-0.4, 1.3, 1], variable = 'x')))
print("Roots: " + str(np.poly1d([-0.4, 1.3, 1], variable = 'x').roots))
```

```python
print("(E)\n" + str(np.poly1d([1.3, 1], variable = 'x')))
print("Roots: " + str(np.poly1d([1.3, 1], variable = 'x').roots))
```


- (F) $X_t = 0.4X_{t-1}+0.45X_{t-2}+\varepsilon_t+\varepsilon_{t-1}+0.25\varepsilon_{t-2}$
- (G) <span style="color:orange">
$X_t = 0.75X_{t-1}-0.5625X_{t-2}+\varepsilon_t+1.25\varepsilon_{t-1}$
</span>
- (H) $X_t = 0.9X_{t-1}+\varepsilon_t$




Consider the process given in (F) $$X_t = 0.4X_{t-1} + 0.45X_{t-2} + \varepsilon_t + \varepsilon_{t-1} + 0.25\varepsilon_{t-2},$$
can be written in operator form as
$$(1 - 0.4B - 0.45B^2)X_t = (1 + B + 0.25B^2)\varepsilon_t.$$
At first, $X_t$ appears to be an $ARMA(2, 2)$ process. But notice that
$$\phi(B) = 1 - 0.4B - 0.45B^2 = (1 + 0.5B)(1 - 0.9B)$$
and
$$\theta(B) = (1 + B + 0.25B^2) = (1 + 0.5B)^2$$
have a common factor that can be canceled. After cancellation, the operators are
$\phi(B) = (1 - 0.9B)$ and $\theta(B) = (1 + 0.5B)$, so the model is an $ARMA(1, 1)$ model,
$(1 - 0.9B)X_t = (1 + 0.5B)\varepsilon_t$, or
$$X_t = 0.9X_{t-1} + 0.5\varepsilon_{t-1} + \varepsilon_t.$$


The model is causal because $\phi(z) = (1 - 0.9z) = 0$ when $z = 10/9$, which is
outside the unit circle.

The model is also invertible because the root of $\theta(z) = (1 + 0.5z)$ is $z = -2$, which is outside the unit circle.


```python
print(r"(F)-$\phi(z)$\n" + str(np.poly1d([-0.45,-0.4, 1], variable = 'z')))
print("Roots are : " + str(np.poly1d([-0.45,-0.4, 1], variable = 'z').roots))
```

```python
print(r"(F)-$\theta(z)$\n" + str(np.poly1d([0.25,1, 1], variable = 'z')))
print("Roots are : " + str(np.poly1d([0.25,1, 1], variable = 'z').roots))
```

```python
print("(F)-$\phi(z)$\n" + str(np.poly1d([-0.45,-0.44, 1], variable = 'z')))

print("Roots are : " + str(np.poly1d([-0.45,-0.44, 1], variable = 'z').roots))
```

We see that for (A), the process is invertible (the roots are outside the unit circle), while in (B) it is not since one of its roots is inside the unit circle. Finally, (E) is also non-invertible.

So, to recap:

- (A) is **stationary** and **invertible**;
- (B) is **stationary** and **non-invertible**;
- (C) is **non-stationary** and **invertible**;
- (D) is **stationary** and **invertible**;
- (E) is **stationary** and **non-invertible**;

We will remind that an MA model is invertible if it is equivalent to a converging infinite order AR model.  By converging, we mean that the AR coefficients decrease to 0 as we move back in time.

## 1.2 Process Simulation

To make it easier to follow, we will simulate each process separately:

(1) $X_t = 1 + \varepsilon_t + 0.5 \varepsilon_{t-1}$

The logic behind generating this process is as follows:

- when $t = 0$, we have $\varepsilon_0 = 0$, $X_0 = 0$
- when $t = 1$, we have $X_1 = 1 + \varepsilon_1 + 0.5 \cdot \varepsilon_0 = 1 + \varepsilon_1$
- when $t = 2$, we have $X_2 = 1 + \varepsilon_2 + 0.5 \cdot \varepsilon_1$
- when $t = 3$, we have $X_3 = 1 + \varepsilon_3 + 0.5 \cdot \varepsilon_2$
- etc.

we can carry out these calculations in `Python` as follows:

```python
np.random.seed(123)
#
N = 150
epsilon = np.random.normal(loc = 0, scale = 0.5, size = N)
#
A_data = 1 + epsilon[0]
for j in range(1, N):
    X_temp = 1 + epsilon[j] + 0.5 * epsilon[j - 1]
    A_data = np.append(A_data, X_temp)
```

(2) $X_t = 1 + \varepsilon_t + 1.3 \varepsilon_{t-1} - 0.4 \varepsilon_{t-2}$

The logic behind generating this process is as follows:

- when $t = 0$, we have $\varepsilon_0 = 0$, $X_0 = 0$
- when $t = 1$, we have $X_1 = 1 + \varepsilon_1 + 1.3 \cdot \varepsilon_0 - 0.4 \cdot \varepsilon_{-1}= 1 + \varepsilon_1$
- when $t = 2$, we have $X_2 = 1 + \varepsilon_2 + 1.3 \cdot \varepsilon_1 - 0.4 \cdot \varepsilon_{0} = 1 + \varepsilon_2 + 1.3 \varepsilon_1$
- when $t = 3$, we have $X_3 = 1 + \varepsilon_3 + 1.3 \cdot \varepsilon_2 - 0.4 \cdot \varepsilon_1$
- when $t = 4$, we have $X_4 = 1 + \varepsilon_4 + 1.3 \cdot \varepsilon_3 - 0.4 \cdot \varepsilon_2$
- etc.

we can carry out these calculations in `Python` as follows:

```python
np.random.seed(123)
#
N = 150
epsilon = np.random.normal(loc = 0, scale = 0.5, size = N)
#
B_data = 1 + epsilon[0]
B_data = np.append(B_data, 1 + epsilon[1] + 1.3 * epsilon[0])
for j in range(2, N):
    X_temp = 1 + epsilon[j] + 1.3 * epsilon[j - 1] - 0.4 * epsilon[j - 2]
    B_data = np.append(B_data, X_temp)
```

We now start the loop from `2` instead of `1`, since the first two elements include $\varepsilon$ at values $t \leq 0$, so they have a different formula.

```python
plt.figure(figsize = (15, 5))
plt.plot(B_data, label = "$X_t = 1 + \\varepsilon_t + 0.5 \\varepsilon_{t-1}$ (invertible)")
plt.plot(A_data, label = "$X_t = 1 + \\varepsilon_t + 1.3 \\varepsilon_{t-1} - 0.4 \\varepsilon_{t-2}$ (non-invertible)")
plt.title("Moving-Average Processes")
plt.legend()
plt.show()
```

Notice that the plots do not exhibit large differences - you may not be able to tell whether a process is invertrible or not from the time series plots alone.

(C) $X_t = 1 + 1.1 X_{t-1} + \varepsilon_t$

The logic behind generating this process is as follows:

- when $t = 0$, we have $\varepsilon_0 = 0$, $X_0 = 0$
- when $t = 1$, we have $X_1 = 1 + 1.1 \cdot X_{0} + \varepsilon_1 = 1 + \varepsilon_1$
- when $t = 2$, we have $X_2 = 1 + 1.1 \cdot X_{1} + \varepsilon_2 = 1 + 1.1 \cdot (1 + \varepsilon_1) + \varepsilon_2$
- when $t = 3$, we have $X_3 = 1+ 1.1 X_{2} + \varepsilon_3 = 1 + 1.1 \cdot (1 + 1.1 \cdot (1 + \varepsilon_1) + \varepsilon_2) + \varepsilon_3$
- etc.

we can carry out these calculations in `Python` as follows:

```python
np.random.seed(123)
#
N = 150
epsilon = np.random.normal(loc = 0, scale = 0.5, size = N)
#
C_data = np.array(1 + epsilon[0], ndmin = 1)
for j in range(1, N):
    X_temp = 1 + 1.1 * C_data[j - 1] + epsilon[j]
    C_data = np.append(C_data, X_temp)
```

(D) $X_t = 1 + 1.1 X_{t-1} - 0.2 X_{t-2} + \varepsilon_t$

The logic behind generating this process is as follows:

- when $t = 0$, we have $\varepsilon_0 = 0$, $X_0 = 0$
- when $t = 1$, we have $X_1 = 1 + 1.1 \cdot X_{0} - 0.2 \cdot X_{-1} + \varepsilon_1 = 1 + \varepsilon_1$
- when $t = 2$, we have $X_2 = 1 + 1.1 \cdot X_{1} - 0.2 \cdot X_0 + \varepsilon_2 = 1 + 1.1 \cdot (1 + \varepsilon_1) + \varepsilon_2$
- when $t = 3$, we have $X_3 = 1+ 1.1 X_{2} - 0.2 X_{1} + \varepsilon_3 = 1 + 1.1 \cdot (1 + 1.1 \cdot (1 + \varepsilon_1) + \varepsilon_2) - 0.2 \cdot (1 + \varepsilon_1)+ \varepsilon_3$
- etc.

we can carry out these calculations in `Python` as follows:

```python
np.random.seed(123)
#
N = 150
epsilon = np.random.normal(loc = 0, scale = 0.5, size = N)
#
D_data = np.array(1 + epsilon[0], ndmin = 1)
D_data = np.append(D_data, 1 + 1.1 * D_data[0] + epsilon[1])
for j in range(2, N):
    X_temp = 1 + 1.1 * D_data[j - 1] - 0.2 * D_data[j - 2] + epsilon[j]
    D_data = np.append(D_data, X_temp)
```

We start the loop from `2` instead of `1`, since the first two elements include pas values of $Y$ at values $t \leq 0$, so they have a different formula.

```python
fig = plt.figure(figsize = (15, 5))
fig.add_subplot(1, 2, 1).plot(C_data)
plt.title("$X_t = 1 + 1.1 X_{t-1} + \\varepsilon_t$ (non-stationary)")
fig.add_subplot(1, 2, 2).plot(D_data)
plt.title("$X_t = 1 + 1.1 X_{t-1} - 0.2 X_{t-2} + \\varepsilon_t$ (stationary)")
plt.show()
```

(E) $X_t = 1.1 X_{t-1} - 0.2 X_{t-2} + \varepsilon_t + 1.3 \varepsilon_{t-1}$

The logic behind generating this process is as follows:

- when $t = 0$, we have $\varepsilon_0 = 0$, $X_0 = 0$
- when $t = 1$, we have $X_1 = 1.1 \cdot X_{0} - 0.2 \cdot X_{-1} + \varepsilon_1 + 1.3 \cdot \varepsilon_0 = \varepsilon_1$
- when $t = 2$, we have $X_2 = 1.1 \cdot X_{1} - 0.2 \cdot X_0 + \varepsilon_2 + 1.3 \cdot \varepsilon_1 = 1.1 \cdot (1 + \varepsilon_1) + \varepsilon_2 + 1.3 \cdot \varepsilon_1$
- when $t = 3$, we have $X_3 = 1.1 X_{2} - 0.2 X_{1} + \varepsilon_3 + 1.3 \cdot \varepsilon_2 = 1.1 \cdot (1 + 1.1 \cdot (1 + \varepsilon_1) + \varepsilon_2 + 1.3 \cdot \varepsilon_1) - 0.2 \cdot (1 + \varepsilon_1)+ \varepsilon_3 + 1.3 \cdot \varepsilon_2$
- etc.

we can carry out these calculations in `Python` as follows:

```python
np.random.seed(123)
#
N = 150
epsilon = np.random.normal(loc = 0, scale = 0.5, size = N)
#
E_data = np.array(epsilon[0], ndmin = 1)
E_data = np.append(E_data, 1.1 * E_data[0] + epsilon[1] + 1.3 * epsilon[0])
for j in range(2, N):
    X_temp = 1.1 * E_data[j - 1] - 0.2 * E_data[j - 2] + epsilon[j] + 1.3 * epsilon[j - 1]
    E_data = np.append(E_data, X_temp)
```

```python
plt.figure(figsize = (15, 5))
plt.plot(E_data, label = "$X_t = 1.1 X_{t-1} - 0.2 X_{t-2} + \\varepsilon_t + 1.3 \\varepsilon_{t-1}$ (stationary non-invertible)")
plt.plot(D_data, label = "$X_t = 1 + 1.1 X_{t-1} - 0.2 X_{t-2} + \\varepsilon_t$ (stationary)")
plt.title("ARMA(2, 1) Process vs AR(2) process")
plt.legend()
plt.show()
```

Note the difference in the mean - our $ARMA$ process does not have an intercept term.

## 1.3 The Mean of the processes

We remember that $\varepsilon_t \sim \mathcal{N}(0, 0.5^2)$.

- (C) $\mathbb{E} (X_t) = \mathbb{E} (1 + 1.1 X_{t-1} + \varepsilon_t) = 1 + 1.1 \mathbb{E} (X_{t-1}) = 1 + 1.1 + (1.1)^2 \mathbb{E} (X_{t-2}) = 1 + 1.1 + (1.1)^2 + (1.1)^3 \mathbb{E} (X_{t-2}) = ... = \sum_{j = 0}^\infty (1.1)^j = \infty$.

Assuming that $X_t = 0$ for $t \leq 0$, this can be expressed as $\mathbb{E} (X_t) =\sum_{j = 0}^{t-1} (1.1)^j$ for an $AR(1)$ process. So, for $t = 150$ this is roughly

```python
tmp_sum = 0
for i in range(0, N):
    tmp_sum = tmp_sum + (1.1)**i
print(tmp_sum)
```

```python
print(C_data.mean())
```

An example of a non-stationary process - a non-constant mean.

- (D) $\mathbb{E} (X_t) = \mathbb{E} (1 + 1.1 X_{t-1} - 0.2 X_{t-2} + \varepsilon_t) = 1 + 1.1 \mathbb{E} (X_{t-1}) - 0.2 \mathbb{E} (X_{t-2})$. For a stationary process $\mathbb{E} (X_t) = \mu$, which gives $\mu = 1 + 1.1 \mu - 0.2 \mu$, which gives $\mu = \dfrac{1}{1 - 1.1 + 0.2} = \dfrac{1}{0.1} = 10$.

```python
print(D_data.mean())
```

- (E) $\mathbb{E} (X_t) = \mathbb{E} (1.1 X_{t-1} - 0.2 X_{t-2} + \varepsilon_t + 1.3 \varepsilon_{t-1}) = 1.1 \mathbb{E} (X_{t-1}) - 0.2 \mathbb{E} (X_{t-2})$. Since $X_t$ is stationary and invertible, we can write it as an infinite MA process. Furthermore, for a stationary process $\mathbb{E} (X_t) = \mu$, which gives $\mu = 1.1 \mu - 0.2 \mu \Longrightarrow \mu = 0$.

## 3 Residual Diagnostics: (3.1), (3.2) and (3.3)

Remember that our underlying assumption is that $\varepsilon_t \sim WN$. In other words the residuals:

- must be zero-mean;
- must have a constant variance;
- must not be autocorrelated;

The zero-mean property is guaranteed by the estimation method (the proof is similar to the OBS estimation method, which minimizes the sum of squared residuals in the univariate cros-sectional regression case). We can examine the variance by checkign for ARCH effects. For the purpose of these exercises, we will examine the variance only visually.

We note that, unlike `forecast` in `R`, the `statsmodels` package in `Python` does not have the `tsdisplay` function. Fortunately, we can always create one ourselves for simplicity:

```python
import pandas as pd
```

```python
def tsdisplay(y, figsize = (14, 8), title = "", lags = 20):
    tmp_data = pd.Series(y)
    fig = plt.figure(figsize = figsize)
    #Plot the time series
    tmp_data.plot(ax = fig.add_subplot(311), title = "$Time\ Series\ " + title + "$", legend = False)
    #Plot the ACF:
    sm.graphics.tsa.plot_acf(tmp_data, lags = lags, zero = False, ax = fig.add_subplot(323))
    plt.xticks(np.arange(1,  lags + 1, 1.0))
    #Plot the PACF:
    sm.graphics.tsa.plot_pacf(tmp_data, lags = 20, zero = False, ax = fig.add_subplot(324))
    plt.xticks(np.arange(1,  lags + 1, 1.0))
    #Plot the QQ plot of the data:
    sm.qqplot(tmp_data, line='s', ax = fig.add_subplot(325))
    plt.title("QQ Plot")
    #Plot the residual histogram:
    fig.add_subplot(326).hist(tmp_data, bins = 40, normed = 1)
    plt.title("Histogram")
    #Fix the layout of the plots:
    plt.tight_layout()
    plt.show()
```

Furthermore, the `tsdiag`, available in `R`, does not have an analog in `statsmodels`. Neverhteless, we can create our own:

```python
def tsdiag(y, figsize = (14, 8), title = "", lags = 30):
    tmp_data = pd.Series(y)
    tmp_data.index += 1
    tmp_acor = list(sm_stat.diagnostic.acorr_ljungbox(tmp_data, lags = lags, boxpierce = True))
    # Plot Bjung-Box and Box-Pierce statistic p-values:
    plt.plot(range(1, len(tmp_acor[0]) + 1), tmp_acor[1], 'bo', label = "Bjung-Box values")
    plt.plot(range(1, len(tmp_acor[0]) + 1), tmp_acor[3], 'go', label = "Box-Pierce values")
    plt.xticks(np.arange(1,  len(tmp_acor[0]) + 1, 1.0))
    plt.axhline(y = 0.05, color = "red", label = "5% critical value")
    plt.title("$Time\ Series\ " + title + "$")
    plt.legend()
    plt.show()
    # Return the statistics:
    col_index = ["Bjung-Box: X-squared", "Bjung-Box: p-value", "Box-Pierce: X-squared", "Box-Pierce: p-value"]
    return pd.DataFrame(tmp_acor, index = col_index, columns = range(1, len(tmp_acor[0]) + 1))
```

# Stationarity and invertibility problems when estimating simulated or empirical data

As we have seen, in case of non-stationary or non-invertible processes, we may still get an estimated model, with stationary and invertible coefficients. The model may not fit the data well, but is an attempt of fitting a *good* model (in terms of stationarity and invertibility). However, sometimes estimation may result in an error, stating that the non-stationarity and/or noninvertibility doesn ot allow to estimate a model.

To make matters worse, sometimes even if we know that a process is stationary and/or invertible, we may still get an error, stating otherwise. To give an example of when an error may occur, generate the following $\rm AR(1)$ process

```python
np.random.seed(1)
N = 100
AR_spec_test = smt.arima_process.ArmaProcess(ar = np.array([1, 0.99]), ma = [1])
AR_bad_example =  AR_spec_test.generate_sample(N, scale = 0.5)
```

We can verify that this process is both stationary and invertible:

```python
AR_spec_test.isstationary
```

- a small sample size;
- a random chance;
- combination of coefficients, which lead to lag function roots that are close to the unit circle;
- a specific lag order selection in the AR, or MA, may result in a non-stationary and/or non-invertible process. **This may be remedied by attempting to specify a different order model.**
- bugs in the econometric software (not as frequent);

#ARIMA

A time series $\{X_t\}$ is said to follow an integrated autoregressive moving average model if the $d^{\text{th}}$ difference $W_t = ∇^dX_t=(1-B)^dX_t$ is a stationary ARMA process. If $\{W_t\}$ follows an $ARMA(p,q)$ model, we say that $\{X_t\}$ is an $ARIMA(p,d,q)$ process. In general, to avoid much complications on calculation and for checking the correctness of the calculations, we can usually take $d = 1$ or at most 2 for practical purposes.

Consider then an $ARIMA(p,1,q)$ process with $W_t = X_t − X_{t − 1}$, we have

$W_t = \phi_1 W_{t-1} + \dots + {\phi_p}W_{t-p}+\varepsilon_t - \{\theta_1 \varepsilon_{t-1} + \dots + \theta_{q} \varepsilon_{t-q}\}$

where $\varepsilon_t \sim WN(0, \sigma^2).$

In terms of the observed data series,

$X_t − X_{t − 1} = \phi_1 \left[X_{t-1} − X_{t − 2}\right] + \dots + {\phi_p}\left[X_{t-p} − X_{t − p-1}\right]+\varepsilon_t - \{\theta_1 \varepsilon_{t-1} + \dots + \theta_{q} \varepsilon_{t-q}\}$

It can be written in terms of backshift operator as,
$\phi(B) (1-B)^d (X_t) = \theta (B) \varepsilon_t $ with $d=1$
