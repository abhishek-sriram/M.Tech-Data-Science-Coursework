**Understanding the linear relationship**

The equation $y=mx+c$ in general case and $y_{t} = c + \phi_{1}y_{t-1}$ in our time series case is linear and the relationship between $x$ and $y$ ( in turn to be $y_{t-1}$ and $y_{t}$ ) variables is also linear. A linear relationship is described as that the increasing or decreasing nature of one variable (which is the independent variable or $x$ or $y_{t-1}$) any number of times, say $n$ times will cause a corresponding increase or decrease of $n$ times in the other variable (which is the dependent variable $y$ or $y_{t}$). In simpler words, if you triple one variable, the other will be tripled as well.

The covariance determine the direction of the linear relationship between two variables. So the direction could be positive, negative and zero.

\begin{equation}
	\gamma(x,y)=\text{Cov}(x,y) = \frac{1}{n-1}\sum\limits_{i=1}^{n}((x_i-\bar{x})(y_i-\bar{y}))
\end{equation}

If we have a multivariate data with dependent variables $x,y$ and $z$ %%depends on independent variable $x$
and we want to compare the covariance between $x$ and $y$  vs $y$ and $z$, what can be understood/implied from the covariance?

For an instance, suppose the covariance of $x$ and $y$ ( $\text{Cov}(x,y)$ ) is a very large value than the covariance of $y$ and $z$ ( $\text{Cov}(y,z)$ ), does it mean that, the two attributes $x$ and $y$ have better linear relationship than the $y$ and $z$ ?

This question may divert our minds to think more on the relation ships between the variables. Let us see the below points for more understanding.

The covariance is actually a product of any two units of data.%% and so, it will be product of two units.
We don't have the information about which units they were representing (for example meters, kilometers, kilograms, liters, \emph{etc.}). Suppose the covariance of $(x,y)$ and $(y,z)$ have different units, we can't compare these two.
It is like comparing two weights whose values are in kilograms and milligrams. In general, they have to be converted to same units before starting any comparison.
Even though this case cannot be dealt by covariance,
we can convert or bring them to unitless/dimensionless variable easily by dividing them by same product of two units.
This can be arrive by dividing the covariance by standard deviation (SD or sd).
The covariance of $(x,y)$ is divided by $sd(x)$ and $sd(y)$  and the $sd(x)$ and $(x_i-\bar{x})$ has same unit. Similarly, $sd(y)$ and $(y_i-\bar{y})$ has same unit. From these we can ultimately introduce the correlation co-efficient (or just correlation between any two variables)

\begin{equation*}
	r(x,y)=\rho(x,y)=\text{Corr}(x,y) = \frac{\text{cov}(x,y)}{\sigma_x\sigma_y} = \frac{\frac{1}{n-1}\sum\limits_{i=1}^{n}((x_i-\bar{x})(y_i-\bar{y}))}{\sigma_x\sigma_y}
\end{equation*}

# The Auto-Regressive Time Series

**Simple Operators**

We shall employ extensively the backward shift operator $B$, which is defined by $Bx_t = x_{t-1}$; it leads to $B^mz_t = x_{t-m}$. The inverse operation is
performed by the forward shift operator $F = B^{-1}$ given by $Fx_t = x_{t+1}$; hence
$F^m x_t = x_{t+m}$. Another important operator is the backward difference operator, $\nabla$,
defined by $\nabla x_t =z_t - x_{t-1}$ . This can be written in terms of $B$, since since
$\nabla x_t = x_t - x_{t-1} = (1 - B)x_t$

A seasonal difference followed by a first difference can be written as
\begin{align*}
	(1-B)(1-B^m)y_t &= (1 - B - B^m + B^{m+1})y_t \\
	&= y_t-y_{t-1}-y_{t-m}+y_{t-m-1},
\end{align*}

**Autoregressive Models**

A stochastic model that can be extremely useful in the representation of certain practically occurring series is the auto-regressive model. In this model, the current value of the process is expressed as a finite, linear aggregate of previous values of the process and a random shock at. Let us denote the values of a process at equally spaced times $t,t - 1, t - 2, \dots$ by $y_{t},y_{t-1},y_{t-2}, \dots$. Also let $\bar{y}_{t} = {y}_{t} - \mu$ be the series of deviations from $\mu$. Then

\begin{align*}
	y_{t} = c + \phi_{1}y_{t-1} + \phi_{2}y_{t-2} + \dots + \phi_{p}y_{t-p} + \epsilon_{t}
\end{align*}

\begin{equation}
	\tilde{y}_{t} = c + \phi_{1}\tilde{y}_{t-1} + \phi_{2}\tilde{y}_{t-2} + \dots + \phi_{p}\tilde{y}_{t-p} + \epsilon_{t}
\end{equation}

If we define an auto-regressive operator of order $p$ in terms of the backward shift operator $B$ by
\begin{equation}
	\phi(B) = 1 - \phi_1B - \phi_2B^2 - \cdots - \phi_pB^p
\end{equation}
the auto-regressive model given in the above equation may be written economically as $\phi(B)y_t = \epsilon_{t}$
The model contains $p + 2$ unknown parameters $\mu, \phi_1, \phi_2, \dots , \phi_p, \sigma_\epsilon^2$, which in practice have to be estimated from the data. The additional parameter $\sigma_\epsilon^2$ is the variance of the white noise process $\epsilon_{t}$.


In general, there are many real world phenomena the past instances/events influences the present happening. %%The events of yesterday can be used to foretell what will happen today.
If we try to express this scenario as a time series, the data are said to have an auto-regressive property. In an auto-regressive time series, the current value can be expressed as a function of its previous value, the value before that one and so forth. In other words, the current value is correlated with previous values from the same time series.

If a time series is auto-regressive it is often the case that the current value's forecast can be computed as a linear function of the previous value and a constant, as follows:
\begin{equation}
	y_{t} = c + \phi_{1}y_{t-1} %%+ \phi_{2}y_{t-2} %%+ \dots + \phi_{p}y_{t-p}
	+ \epsilon_{t}
\end{equation}

For an AR(1) model
\begin{equation}
	y_{t} = c + \phi_{1}y_{t-1} + \epsilon_{t},
\end{equation}
we have the following cases.

We normally restrict autoregressive models to stationary data, in which case some constraints on the values of the parameters are required.

For an AR(1) model: $-1<\phi_1<1$

For an AR(2) model: $-1<\phi_1<1, \phi_1+\phi_2<1$,

For $p\geq 3$, the calculation of restrictions are little bit tedious, so that in those cases, programming help is required. Most of them will use the programming languages/softwares like *R*, *Python* etc. for the calculation of restrictions when estimating the parameters of a higher order model.


In AR(1), $y_{t}$ is the value that is forecast by the equation at the time step $t$. $c$ is the $Y$-intercept (here $Y$ represents the $Y-$ axis) of the model and it applies a constant amount of bias on the forecast. It also specifies what will be the forecast for $y_{t}$ if the value at the previous time step $y_{t-1}$ happens to be zero. $\phi_{1}$ tells us the rate at which $y_{t}$ increases/decreases (changes) based on the influence of $y_{t-1}$.

The main assumption upon this simple equation is that the variance in $y_{t-1}$ is able to explain all the variance expressed by all values that are older than $y_{t-1}$ in the time series. It is as if $y_{t-1}$ captures all the information associated with values older than itself.

On the other hand, what happens if this assumption were not true/satisfying? Also, What happens when the variance in $y_{t-1}$ is not able to explain all of the variance covered within $y_{t-2}$? In that case, the above equation will not be able to feed this unexplained portion of the variance from $y_{t-2}$ into $y_{t}$, causing the forecast for $y_{t}$ to go off the mark.

Fortunately it's easy to fix this problem adding a term to the above equation as follows:
\begin{equation}
	y_{t} = c + \phi_{1}y_{t-1} + \phi_{2}y_{t-2} %%+ \dots + \phi_{p}y_{t-p}
	+ \epsilon_{t}
\end{equation}
In this equation, the added term $\phi_{2}y_{t-2}$ tries (seeks) to capture the variance contained/covered in values that are older than $y_{t-1}$ that could not be explained by the variance in $y_{t-1}$. It feeds this balance amount of information directly into the forecast for todays value $y_{t}$.

## ACF - Autocorrelation Function

This type of correlation is used to understand how the time series observations depend on with values of the same data series at previous time data point.
The past observations in the series are called as lags of the series data.
Consider the dataset below:

| Original Data   | Lag-1 | Lag-2 | Lag-3 |
|-----------------|-------|-------|-------|
| 1               |       |       |       |
| 2               | 1     |       |       |
| 3               | 2     | 1     |       |
| 4               | 3     | 2     | 1     |
| 5               | 4     | 3     | 2     |

In the above dataset, the first column is an original data time series, and the second column is a one time step shift original time series data and its called as Lag-1 of the time series original data. The length of this Lag-1 series will be one less than the original time series. Similarly we can compute Lag-2, Lag-3, *etc.*

Now let us get into the mathematical format to compute the ACF. We can compute the correlation coefficient between original and Lag-1 time series.

Let $L_1$ =Lag-1 of $x$ time series. The length of $L1$ series will be $n-1$. Since the summation is applied for both the series, the summation will have $n-1$ terms.
\begin{align*}
	\text{Corr}(x_t,L_1) = \frac{\text{cov}(x_t,L_1)}{\sigma_x\sigma_{L_1}} = \frac{\frac{1}{n-1}\sum_{i=1}^{n-1}((x_{i+1}-\bar{x})(L_{1i}-\bar{x}))}{\sigma_x\sigma_{L_1}} = \text{ACF}(\text{Lag-1})
\end{align*}
Note that we are ignoring the difference between $\overline{x}$ and $\overline{L_1}$

We just now computed the ACF of Lag-1, and the same we can extend to generalize the lag terms.

The value of ACF($L_k$) is called the autocorrelation coefficient at lag k. The plot of the sample autocorrelations ACF($L_k$) versus $k$ (the time lags) is called the correlogram or autocorrelation plot.

The correlogram is a widely used tool for identifying the randomness in a data set. This randomness is ascertained by computing auto-correlations for data values at varying time lags. If any randomness occurs, such auto-correlations should be near zero for any and all time-lag separations. If there is no randomness occurs, then one or more of the auto-correlations will be significantly non-zero.

## ACVF - Autocovariance Function

For a given random process, the auto-covariance is a function that gives the covariance of the process with itself at pairs of time points.
Let $Y$ be the data series and each observations are collected at regular intervals of time. The expectation of $Y$ is given as $\mu=E[Y]$. Subtracting the $\mu$ from each observations of $Y$ gives the $Z$.



Multiplying the Vector $Z$ with $Z$ transpose, gives the $(N \times N)$ matrix as below.

\begin{eqnarray*}
ZZ^T &=&
		\left[\begin{array}{cc}
			Y_1-\mu \\
			Y_2-\mu\\
			Y_3-\mu \\
			\vdots \\
			Y_n-\mu \\
		\end{array}
		\right]
	\left[\begin{array}{cc}
			Y_1-\mu, Y_2-\mu, Y_3-\mu, \dots, Y_n-\mu \\
		\end{array}
		\right] \\
				%%\left [
&=&		{\scriptsize
	\begin{bmatrix}
			(Y_1-\mu)^2 & (Y_1-\mu)(Y_2-\mu) &  (Y_1-\mu)(Y_3-\mu)&\dots &(Y_1-\mu)(Y_n-\mu) \\
			(Y_2-\mu)(Y_1-\mu)&(Y_2-\mu)^2 & (Y_2-\mu)(Y_3-\mu) & \dots &(Y_2-\mu)(Y_n-\mu)\\
			(Y_3-\mu)(Y_1-\mu)& (Y_3-\mu)(Y_2-\mu) & (Y_3-\mu)^2  &\dots & (Y_3-\mu)(Y_n-\mu) \\
			\vdots & \vdots &\vdots &\vdots &\vdots \\
			(Y_n-\mu)(Y_1-\mu) & (Y_n-\mu)(Y_2-\mu)& (Y_n-\mu)(Y_3-\mu) & \dots & (Y_n-\mu)^2 \\
		\end{bmatrix} }
		%%\right]	}$
\end{eqnarray*}


So the expectation of the above matrix gives the covariance matrix.
$E[ZZ^T]= \text{cov}(Y,Y) = E[(Y-\mu)(Y-\mu)^T]=E[YY^T] - \mu\mu^T.	$
The $E[Z Z^T]$ matrix is a variance-covariance matrix, which is a square matrix that contains the variances and covariances associated with several variables. The diagonal elements of the matrix contain the variances of the variables and the off-diagonal elements contain the covariances between all possible pairs of variables.

The auto covariance function (ACVF) at lag-$k$ for the time series is defined by
\begin{align*}
	\gamma_k = \frac{\frac{1}{n-1}\sum_{i=1}^{n-k}((y_i-\bar{y})(y_{i+k}-\bar{y}))}{\sigma_x\sigma_y}
\end{align*}
Easily, the ACVF for various lags can be computed from the variance-covariance matrix. In the matrix, the values on the main diagonal are $\gamma_0$, the values on the diagonal above and below the main diagonal are $\gamma_1$. The values on the diagonal two units away are $\gamma_2$ and so on.

Let us compute the auto covariance function (ACVF) for the given data set. A simple python code to compute the variance-covariance matrix.


```python
# data = [27, 28, 29, 30, 32, 32, 33]
data=[200,]
##(Create a Matrix $Z$ for the data set)
Z = np.array(data)
print('Z=',Z)
print('=======================================================')
##  (Compute the Mean and substract it from the data)
meanZ = Z.mean()
print('mean of Z=',meanZ)
print('=======================================================')
Z1 = Z - meanZ
print('Z1=',Z1)
print('=======================================================')
## (Transpose the Z matrix and perform the mat multiplication Z and Z')
Ztrans = np.transpose(Z1)
print('Z Transpose=',Ztrans)
print('=======================================================')
rho = np.dot(Z1[:,None],Ztrans[None,:])
print('rho=',rho)
print('=======================================================')
```

From the result of $n\times n$ matrix, let us compute the auto covariance function (ACVF) values as given below.


```python
l = len(data)
slist = []
for i in range(0,l):
    s = 0
    for j in range(0,l-i):
        s = s + rho[j+i][j]
    slist.append(s)
    print('s=',s)
    print('=======================================================')
    print('slist=',slist)
    print('=======================================================')

print('slist=',slist)
print('=======================================================')
plt.plot(range(0,l), slist)
plt.show()
```

# Partial AutoCorrelation Function (PACF)

Many of us understand that the auto-correlation function as *correlation between the observation at the current time and the observations at previous times.*
The partial auto-correlation function (PACF) is also to be understood in other way like a correlation between the observation at the current time and the observation at previous time given that it has to be considered both observations are correlated to inputs/observations at other times.
To brief this a little, consider an example, today 30-Apr stock price can be correlated to the 28-Apr stock price and yesterday 29-Apr stock price can also be correlated to the 28-Apr. Then PACF of yesterday 29-Apr stock price is the real correlation between today 30-Apr and 29-Apr after taking out the influence of the 28-Apr stock price.



PACF_diag.png

For an instance, consider another example, let us think we have input variables as $x_1$, $x_2$ and $x_3$ and an output variable as $y$.
The PACF between $y$ and $x_3$ is the correlation between the variables $y$ and $x_3$ determined taking into account how both $y$ and $x_3$ are related to $x_1$ and $x_2$.

To understand how $y$ and $x_3$ are related to $x_1$ and $x_2$, we need to compute/perform two regressions as given below.

<!-- \begin{itemize}[label=$\star$]

\item Regression-2 : We have to compute/predict $x_3$ from $x_1$ and $x_2$ by writing an equation by making $x_3$ as dependent variable in terms of $x_1$ and $x_2$.
\end{itemize} -->

<ul>
<li>
Regression-1 : We have to compute/predict $y$ from $x_1$ and $x_2$ by writing an equation by making $y$ as dependent variable in terms of $x_1$ and $x_2$.
</li>
<li>
Regression-2 : We have to compute/predict $x_3$ from $x_1$ and $x_2$ by writing an equation by making $x_3$ as dependent variable in terms of $x_1$ and $x_2$.
</li>
</ul>


If we note there is/are a/few residual/s left out in each regression.
The residuals are not going to be explained by the input variables $x_1$ and $x_2$.
Computing the correlation between these residuals will give the partial correlation between the variables $y$ and $x_3$.

The correlation between $y$ and $x_3$ is computed as below and it computes the full correlation consider influences of others variables.

\begin{align*}
	\rho(y,x_3)=\frac{\gamma(y,x_3)}{\sigma_y\sigma_{x_3}}
\end{align*}

The PACF computes the correlation between $y$ and $x_3$ while removing the influence factors of others variables.

\begin{align*}
	PACF(y,x_3)=\frac{\gamma(y,x_3\mid x_1,x_2)}{\sigma_{y\mid x_1,x_2}\sigma_{x_3\mid x_1,x_2}}
\end{align*}
For a time series, the $k^{th}$ order PACF is the partial correlation of $y_i$ with $y_{i-k}$ condition on $y_{i-1},\dots,y_{i-k+1}.$ \emph{i.e.}

\begin{align*}
	PACF(y,y_{i-k})=\frac{\gamma(y_i,y_{i-k} \mid y_{i-1},y_{i-2},\dots,y_{i-k+1})}{\sigma_{y_i\mid y_{i-1},y_{i-2},\dots,y_{i-k+1}}\sigma_{x_3 \mid x_1,x_2}}
\end{align*}



## PACF Computation

For the given data set, let us compute the PACF for Lag-2. For that the below equation is to be solved.
\begin{align*}
	PACF(y,y_{i-2})=\frac{\gamma(y_i,y_{i-2} \mid y_{i-1})}{\sigma_{y_i\mid y_{i-1}}\sigma_{y_{i-k}\mid y_{i-1}}}
\end{align*}

In this stage we all have to follow the \textbf{Yule-Walker} equations to compute the PACF values. At high level, the method looks as below.

Here we have a simple python code to compute the PACF for various lags.
Let us compute the PACF for Lag-5.


```python
lag = 5
acvf_lags = slist[1:lag+1]
matr = np.zeros((lag, lag))
for i in range(0,lag) :
    for j in range(0,lag-i) :
        matr[j+i][j] = slist[i]
        matr[j][j+i] = slist[i]
        print('matr=',matr)
        print('=======================================================')
        print('slist=',slist)
        print('=======================================================')
ainv = np.linalg.inv(matr)
result1 = np.matmul(ainv,acvf_lags)

print('a inverse=',ainv)
print('=======================================================')
print('result1=',result1)
print('=======================================================')
plt.plot(range(lag), result1)
plt.show()
```

Autocorrelation Function (ACF) and Partial Autocorrelation Function (PACF) are essential tools in time series analysis to understand the correlation structure within a time series data. They help identify patterns in the data and assist in determining the order of autoregressive (AR) and moving average (MA) terms in ARIMA modeling.

Here's how you can manually calculate ACF and PACF with Python code examples using libraries like NumPy and statsmodels:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Generate some example time series data
np.random.seed(0)
n = 100
time = np.arange(n)
data = np.random.normal(loc=0, scale=1, size=n)

# Plot the time series data
plt.figure(figsize=(10, 4))
plt.plot(time, data)
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Example Time Series Data')
plt.grid(True)
plt.show()

# Manual calculation of ACF
def autocorr(x, lag):
    n = len(x)
    mean = np.mean(x)
    num = np.sum((x[:n-lag] - mean) * (x[lag:] - mean))
    denom = np.sum((x - mean) ** 2)
    return num / denom

lags = range(1, 21)  # Calculate ACF for lags 1 to 20
acf_values = [autocorr(data, lag) for lag in lags]

# Plot ACF
plt.figure(figsize=(10, 4))
plt.bar(lags, acf_values)
plt.xlabel('Lag')
plt.ylabel('ACF')
plt.title('Autocorrelation Function (ACF)')
plt.grid(True)
plt.show()

# # Manual calculation of PACF using Yule-Walker equations
# def pacf(x, lag):
#     if lag == 0:
#         return 1.0
#     n = len(x)
#     r = [autocorr(x, l) for l in range(lag + 1)]
#     R = np.array([[r[abs(i - j)] for i in range(lag + 1)] for j in range(lag + 1)])
#     r = np.array(r[1:])
#     phi = np.linalg.solve(R, r)
#     return phi[-1]

def pacf(x, lag):
    if lag == 0:
        return 1.0
    n = len(x)
    r = [autocorr(x, l) for l in range(lag + 1)]
    R = np.zeros((lag + 1, lag + 1))
    for i in range(lag + 1):
        for j in range(lag + 1):
            R[i, j] = r[abs(i - j)]
    r = np.array(r[1:])
    phi = np.linalg.solve(R[1:, 1:], r)
    return phi[-1]


pacf_values = [pacf(data, lag) for lag in lags]

# Plot PACF
plt.figure(figsize=(10, 4))
plt.bar(lags, pacf_values)
plt.xlabel('Lag')
plt.ylabel('PACF')
plt.title('Partial Autocorrelation Function (PACF)')
plt.grid(True)
plt.show()

# Using statsmodels to plot ACF and PACF
plot_acf(data, lags=20)
plt.title('Autocorrelation Function (ACF) using statsmodels')
plt.show()

plot_pacf(data, lags=20)
plt.title('Partial Autocorrelation Function (PACF) using statsmodels')
plt.show()
```

#Autocorrelation Function (ACF):
The Autocorrelation Function (ACF) measures the correlation between a time series and its lagged values. It is calculated for various lags (time intervals), indicating how correlated the series is with itself at different time spans.

##Key Points:

####Range:
The ACF plot typically extends to a certain number of lags.

Interpretation: Peaks in the ACF plot indicate significant correlations between the series and its lagged values at those lags. If a peak extends beyond the shaded region (confidence interval), it suggests a statistically significant correlation.

####Pattern:
The ACF plot can help identify seasonality in the data. For instance, if there's a repeating pattern in the peaks every 7 lags, it may indicate weekly seasonality.

Partial Autocorrelation Function (PACF):
The Partial Autocorrelation Function (PACF) measures the correlation between a time series and its lagged values, while controlling for the effect of other lags in between. It helps to identify the direct relationship between observations at different time spans.

####Key Points:

####Controlled Correlation:
Unlike ACF, PACF removes the correlation contributed by the intermediate lags, focusing only on the direct relationship between the current observation and the lagged observation being considered.

####Identification of AR Terms:
In the context of ARIMA modeling, PACF is particularly useful for determining the order of the autoregressive (AR) component. Significant spikes in PACF beyond the confidence interval at certain lags indicate the number of significant AR terms.

####Interpretation:
Significant spikes in the PACF plot indicate the direct effect of the corresponding lag on the current observation, after accounting for the effects of other lags. These spikes help identify the order of the AR process.

###Differences:
####Scope:
ACF measures the correlation at all lags, while PACF measures the correlation at each lag after removing the effects of the intermediate lags.

####Interpretation:
ACF helps identify overall correlation patterns and seasonality, while PACF helps identify the direct relationship between specific lags, aiding in determining the AR order in time series modeling.

####Usage:
ACF and PACF are often used together in time series analysis. ACF provides a broad overview of the correlation structure, while PACF assists in identifying the specific lag structure for building ARIMA models.

\------------------------------------


While ACF and PACF serve similar purposes of exploring autocorrelation in time series data, their approaches and interpretations differ. ACF provides a general overview, while PACF helps pinpoint the direct relationships between individual observations at different lags, aiding in model specification.
